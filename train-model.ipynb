```python
from sklearn.datasets import load_digits
#importing the load_digits method from datasets sub module in sklearn module

dataset=load_digits()

data=dataset.data
images=dataset.images
target=dataset.target
```

```python
import numpy as np

array=np.random.randint(0,100,(5,3,2))
print(array)
```

```python
from matplotlib import pyplot as plt

plt.imshow(images[100],cmap='gray')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

train_data,test_data,train_target,test_target=train_test_split(data,target,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier

algorithm=KNeighborsClassifier(n_neighbors=3)

algorithm.fit(train_data,train_target)
```

```python
predicted_targets=algorithm.predict(test_data)
```

```python
from sklearn.metrics import accuracy_score

accuracy=accuracy_score(test_target,predicted_targets)
print('Accuracy: ',accuracy)
```
